## Problem 4.1

Suppose that X is a discrete random variable with probability mass function given by

$p(1) = \frac{1}{10}$, $p(2) = \frac{3}{10}$,  $p(3) = \frac{2}{10}$, $p(4) = \frac{3}{10}$, and $p(5) = \frac{1}{10}$

PMF:

```python
import matplotlib.pyplot as plt
import math
import numpy as np

values = [1,2,3,4,5]
probs = [.1, .3, .2, .3, .1]

plt.bar(values, probs)

```

CDF:

$F(1) = .1$

$F(2) = .1 + .3 = .4$

$F(3) = .1 + .3 + .2 = .6$

$F(4) = .1 + .3 + .2 + .3 = .9$

$F(5) = .1 + .3 + .2 + .3 = 1$


```python

cumulative = [.1, .4, .6, .9, 1]

plt.bar(values, cumulative)
```

```python
expectation = 0
for i in range(len(values)):
    expectation += probs[i]*values[i]    
print("E(x): ", expectation)
    
variance = 0
for value, prob in zip(values,probs):
    variance += prob * (expectation - value)**2
    
    
print("Var(x):", variance)
    
```

$\begin{gather*}E[s^2(n)] = E[\frac{(x_1 - \bar{x})^2 + (x_2 - \bar{x})^2 ... + (x_n - \bar{x})^2}{n-1}] = \\
